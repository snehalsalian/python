```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
fig,ax=plt.subplots()
df['job_title_short'].value_counts().plot(kind='bar')
```

```python
df['job_schedule_type'].value_counts().head(3)
```

```python
fig,ax=plt.subplots(1,2)

df['job_title_short'].value_counts().plot(kind='bar',ax=ax[1])
df['job_schedule_type'].value_counts().head(3).plot(kind='bar',ax=ax[0])

#makes layout more tidy without any overlapping 
fig.tight_layout()


 
```

```python
df_skills = df.copy()
df_skills = df_skills.explode('job_skills')
skills_count = df_skills.groupby(['job_skills', 'job_title_short']).size()
df_skills_count = skills_count.reset_index(name='skill_count')
df_skills_count.sort_values(by='skill_count', ascending=False, inplace=True)

df_skills_count
```

```python
job_titles=['Data Analyst','Data Scientist','Data Engineer']
fig,ax=plt.subplots(3,1)

for i,job in enumerate(job_titles):    #enumerate creates index values as well
    df_plot=df_skills_count[df_skills_count['job_title_short']==job].head(5)
    df_plot.plot(kind='barh',y='skill_count',x='job_skills',ax=ax[i],title=job)     #On a barh x means vertical values
    ax[i].invert_yaxis()
    ax[i].legend().set_visible(False)
    ax[i].set_ylabel("")
    ax[i].set_xlim(0,120000)
    

fig.suptitle("Top 5 Skills for Data Jobs",fontsize=15)
fig.tight_layout()                 #Put this below,else subtitle will overlap
    
    
```
