```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df_IN=df[df['job_country']=='India'].copy()
df_IN
```

```python
df_IN['job_posted_month']=df_IN['job_posted_date'].dt.strftime('%B')
df_IN
```

```python
df_IN_pivot=df_IN.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
```

```python
df_IN_pivot.reset_index(inplace=True) 
df_IN_pivot['month_no']=pd.to_datetime(df_IN_pivot['job_posted_month'],format='%B').dt.month
df_IN_pivot['month_no']
```

```python
df_IN_pivot.set_index('index',inplace=True)
```

```python
df_IN_pivot.sort_values('month_no',inplace=True)
df_IN_pivot.set_index('job_posted_month',inplace=True)
df_IN_pivot.drop(columns='month_no',inplace=True)
```

```python
df_IN_pivot
```

```python
df_IN_pivot.plot(kind='line')
```

```python
top3=df_IN['job_title_short'].value_counts().head(3)
top3=top3.index.tolist()    #so that it extracts the index i,e that job names and we convert it to list  to be able to use below
df_IN_pivot[top3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in India')
plt.xlabel('2024')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
